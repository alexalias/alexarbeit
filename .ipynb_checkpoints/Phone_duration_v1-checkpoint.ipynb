```python
# Returns a dictionary of phoneme occurences (keys) in the training data and their durations (values), and LSR (values)
# Looks like {a: [1452, 0.8, 799, 0.5], b : [655, 0.5, 799, 0.45]...}
def read_trainig_files():
	training_dict = defaultdict(list)
	os.chdir("C:/Users/alexutza_a/Abschlussarbeit/DB_Verbmobil/Evaluation/Training")
	
	#Iterate over the training files
	for file in glob.glob("*.par"):
		work_file = open(file)

		for line in work_file:
			if re.match("MAU", line):
				training_dict[line.split()[4]].append(int(line.split()[2]))
				word_duration, phon_count, syl_count = speech_rate.word_duration(file, int(line.split()[3]))
				training_dict[line.split()[4]].append((word_duration/0.0000625)/phon_count) # speech rate as time/phon
				#training_dict[line.split()[4]].append(speech_rate.local_speech_rate(file, int(line.split()[3])))

		work_file.close()
	
	# Remove breaks from the data
	x = training_dict.pop("<p:>")

	return training_dict
#read_trainig_files()
```

```python
# A dictionary giving the values of Mean and SD in a list for each encountered phoneme.
def phone_stats(training_dict):
	stat_dict = defaultdict(list)

	for phoneme in training_dict.keys():
		stat_dict[phoneme].append(int(round(np.mean(training_dict[phoneme][::2]), 0)))
		stat_dict[phoneme].append(int(round(np.std(training_dict[phoneme][::2]), 0)))

	return stat_dict
```

```python
# Returns a list of phoneme occuring in the test files, followed by their respective durations
# Looks like: ["a", 583, 0.5, "b", 12, 0.78, "a", 489, 0.12, ...]
def read_testfiles():
	compare_list = []
	os.chdir("C:/Users/alexutza_a/Abschlussarbeit/DB_Verbmobil/Evaluation/Test")

	#Iterate over the test files
	for file in glob.glob("*.par"):
		work_file = open(file)

		for line in work_file:
			if re.match("MAU", line):
				compare_list.append(str(line.split()[4]))
				compare_list.append(int(line.split()[2]))
				word_duration, phon_count, syl_count = speech_rate.word_duration(file, int(line.split()[3]))
				#compare_list.append((word_duration/0.0000625)/phon_count) # speech rate as word_duration / # phonemes
				compare_list.append(speech_rate.local_speech_rate(file, int(line.split()[3])))
		work_file.close()
	#print(compare_list)
	# Remove breaks from the data

	# Get list of indexes for occurences of <p:>
	pause_index = [i for i, val in enumerate(compare_list) if val == "<p:>"]
	#print(len(pause_index))
	pause_dur = [i + 1  for i in pause_index]
	pause_stat = [j + 1  for j in pause_dur]
	p_l = [x for y in zip (pause_index, pause_dur) for x in y]
	p_list = p_l + pause_stat
	#print(len(p_list))
	actual_list = []

	#print(compare_list)
	ind = 0
	# Copy list to new list, without pauses
	for el in compare_list:
		if ind not in p_list:
			actual_list.append(el)
		ind += 1

	return actual_list
```

```python
# Quote of phoneme mean duration being greater than the local speech rate (as word duration / no. of phonemes)
def test_mean(training_dict, stat_dict):
	test_dict = defaultdict(float)

	for elem in training_dict.keys():
		rate_list = training_dict[elem][1::2]
		test_mean_list = [ 1 for x in rate_list if x <= stat_dict[elem][0]]
		test_dict[elem] = len(test_mean_list)/len(rate_list)
	return test_dict
#DataFrame.from_dict(test_mean(read_trainig_files(), phone_stats(read_trainig_files())), orient="index")
```

```python
# Create a list of speech rates and a list of word durations from the training data
def dur_vs_rate(training_dict):
    duration_list = []
    rate_list = []
    for el in training_dict.keys():
        duration_list += training_dict[el][::2]
        rate_list += training_dict[el][1::2]
    return duration_list, rate_list
duration_list, rate_list = dur_vs_rate(read_trainig_files())

# Calculate the Pearson corrCoef for phoneme duration vs speech rate as word_dur / no. of phones
print("Pearson correlation coefficient:" + str(pearsonr(duration_list, rate_list)[0]))

# Plot phoneme duration vs. speech rate as word_dur / no. of phones
plt.scatter(duration_list, rate_list)
plt.xlabel("Word durations")
plt.ylabel("Speech rate (w_dur/no. of phon)")
plt.show()
```

```python
# Sort value list of the word_dict
from operator import itemgetter
def sort_w_dict(word):
    l = word_dict[word]
    l_gr = [l[i:i+3] for i in range(0, len(l),3)] # transform list in list of 3-elem-lists
    l_sorted = sorted(l_gr,key=itemgetter(1))
    lf_sorted=[]
    for i in range(len(l_sorted)):    # flatten sorted list
        lf_sorted += l_sorted[i]
    return lf_sorted
print(sort_w_dict("Arbeitstreffen"))
```

```python
# Two lists (word_durations_list and phoneme_durations_in_word as list of lists) for given list of words
def multiple_word_stats(word_list):
    w_duration_list = []
    w_sounds_lists = []
    for word in word_list:
        w_duration_list = sort_w_dict(word)[1::3]
        w_sounds_lists = sort_w_dict(word)[2::3]
    return w_duration_list, w_sounds_lists
```

```python
# List of proportions of given phoneme in "genau", expressed as phon_dur/word_dur
def get_genau_propLists(phoneme):
    sound_dur_list = wordSound_dict[phoneme][::2]
    w_dur_list = wordSound_dict[phoneme][1::2]
    proportions_list = []
#    print(phoneme + " " + str(min(sound_dur_list)))
#    print(phoneme + " " + str(np.median(sound_dur_list)))
#    print(phoneme + " " + str(np.mean(sound_dur_list)))
#    print(phoneme + " " + str(max(sound_dur_list)))
    for el in sound_dur_list:
        proportions_list.append(round(el/w_dur_list[sound_dur_list.index(el)], 2))
    return proportions_list
```

```python
# Creates a dict of phoneme proportions in words of specific lengths (as no of phonemes) - w_dur in msec.
# Looks like: {"a" : {1pho_w : [w_dur1, pho_dur1, pho_prop1, w_dur2, pho_dur2, pho_prop2, ...], 2pho_w : [...], ...}, 
#              "b" : {1pho_w : [w_dur1, pho_dur1, pho_prop1, w_dur2, pho_dur2, pho_prop2, ...], 2pho_w: [...], ...}, 
#               ...}

def build_simple_pdict():
    simple_phoprop_dict = defaultdict(dict) # the dict to be returned
    simple_prop_dict = defaultdict(list)   # the value dict
    #os.chdir("C:/Users/alexutza_a/Abschlussarbeit/DB_Verbmobil/Evaluation/Training")
    
    #Iterate over the training files
    for datei in path_list:
        work_file = open(datei)
        for line in work_file:
            if re.match("MAU", line):
                w_dur, phon_count, syl_count = speech_rate.word_duration(datei, int(line.split()[3]))
                key = str(phon_count)+"pho_w"
                simple_prop_dict[key].append(round(w_dur*0.0625, 2))                 # w_dur used in msec.
                simple_prop_dict[key].append(round(int(line.split()[2])*0.0625, 3))   # pho duration in msec.
                simple_prop_dict[key].append(round(int(line.split()[2])/w_dur, 3))   # calc pho prop as pho_dur/w_dur
                if key in simple_phoprop_dict[line.split()[4]].keys():
                    simple_phoprop_dict[line.split()[4]][key] += simple_prop_dict[key]
                else: 
                    simple_phoprop_dict[line.split()[4]][key] = simple_prop_dict[key]
                simple_prop_dict.clear()
        work_file.close()
    return simple_phoprop_dict
```

```python
# Build the simple phoneme dictionary with word lengths
%time simple_pdict = build_simple_pdict()
```

```python
# Help function for building the list of words
# Gets the actual word from file (ORT tier)
def get_word(datei, word_no):
    work_file = open(datei)
    word = ""
    for line in work_file:
        if re.match("ORT", line) and (int(line.split()[1]) == word_no):
            word = str(line.split()[2])
    work_file.close()
    return word
            
```

```python
# Build a list of words containing a specific phoneme, and having a particular length (no. of elem.)
def get_word_list(phoneme, w_length):
    w_list = []
    for datei in path_list:
        work_file = open(datei)
        for line in work_file:
            if re.match("MAU", line) and (str(line.split()[4]) == phoneme):
                #print("aha")
                w_dur, phon_count, syl_count = speech_rate.word_duration(datei, int(line.split()[3]))
                if w_length == phon_count:
                    w_list.append(get_word(datei, int(line.split()[3])))
        work_file.close()
    return w_list
```

```python
from operator import itemgetter

# sort needed dict_list from the simple (!) dict
def sort_dict_val(dict_value_list):
#    dict_value_list = simple_pprop_dict["a:"]["2pho_w"]
    l_gr = [dict_value_list[i:i+3] for i in range(0, len(dict_value_list),3)] # transform list in list of 3-elem-lists
    l_sorted = sorted(l_gr,key=itemgetter(0)) # sort list of of 2-elem-lists
    org_simpleDict_val = []
    for i in range(len(l_sorted)):    # flatten sorted list
        org_simpleDict_val += l_sorted[i]
    return org_simpleDict_val
#print(sort_dict_val(simple_pprop_dict["a:"]["2pho_w"])[:14])
```

```python
# Creates a dict of phoneme proportions in words of specific lengths (as no of phonemes) - w_dur in msec.
# Looks like: {"a" : {1 : [w_dur1, pho_dur1, pho_prop1, w_dur2, pho_dur2, pho_prop2, ...], 2: [...], ...}, 
#              "b" : {1 : [w_dur1, pho_dur1, pho_prop1, w_dur2, pho_dur2, pho_prop2, ...], 2: [...], ...}, 
#               ...}

def build_int_pdict():
    simple_phoprop_dict = defaultdict(dict) # the dict to be returned
    simple_prop_dict = defaultdict(list)   # the value dict
    #os.chdir("C:/Users/alexutza_a/Abschlussarbeit/DB_Verbmobil/Evaluation/Training")
    
    #Iterate over the training files
    for datei in path_list:
        work_file = open(datei)
        for line in work_file:
            if re.match("MAU", line):
                w_dur, phon_count, syl_count = speech_rate.word_duration(datei, int(line.split()[3]))
                key = phon_count
                simple_prop_dict[key].append(round(w_dur*0.0625, 2))                 # w_dur used in msec.
                simple_prop_dict[key].append(round(int(line.split()[2])*0.0625, 3))   # pho duration in msec.
                simple_prop_dict[key].append(round(int(line.split()[2])/w_dur, 3))   # calc pho prop as pho_dur/w_dur
                if key in simple_phoprop_dict[line.split()[4]].keys():
                    simple_phoprop_dict[line.split()[4]][key] += simple_prop_dict[key]
                else: 
                    simple_phoprop_dict[line.split()[4]][key] = simple_prop_dict[key]
                simple_prop_dict.clear()
        work_file.close()
    return simple_phoprop_dict
```

```python
# Build the simple phoneme dictionary with word lengths
%time int_pdict = build_int_pdict()
```

```python
# Create 3 lists for evaluating phoneme durations and proportions for diff w_lengths
def get_coord_lists(phoneme, w_length):
    wdur_mslist = sort_dict_val(int_pdict[phoneme][w_length])[::3]   # xpho-word duration in msec
    wdur_slist = [ round(x/1000, 3) for x in wdur_mslist]     # xpho-word duration in sec
    pdur_list = sort_dict_val(int_pdict[phoneme][w_length])[1::3]    # phoneme duration list (in xphon-words)
    prop_list = sort_dict_val(int_pdict[phoneme][w_length])[2::3]    # list of phoneme proportions in xpho-words
    return wdur_mslist, wdur_slist, pdur_list, prop_list
```

```python
# Help function to calculate w_leng at which phoneme_prop_steak gets below certain values
def threshold_wLeng(phoneme):
    w_leng = sorted(int_pdict[phoneme].keys())
    Q1_list, Q3_list, Median_list, Mean_list = [], [], [], []
    for el in w_leng:
        wdur_mslist, wdur_slist, pdur_list, prop_list = get_coord_lists(phoneme, el)
        Q1_list.append(round(np.percentile(prop_list, 25),2))
        Q3_list.append(round(np.percentile(prop_list, 75),2))
        Median_list.append(round(np.median(prop_list), 2))
        Mean_list.append(round(np.mean(prop_list), 2))
    steak_list = [ y - x for x,y in zip(Q1_list, Q3_list)]
    med02, m02, med01, m01 = 0, 0, 0, 0
    for el in Median_list:
        if el <= 0.2:
            med02 = Median_list.index(el)
            for el in Median_list[med02:]:
                if el < 0.1:
                    med01 = med02 + Median_list[med02:].index(el) +2
                    break
            break
                    
    for el in Mean_list:
        if el <= 0.2:
            m02 = Mean_list.index(el)
            for el in Mean_list[m02:]:
                if el < 0.1:
                    m01 = m02 + Mean_list[m02:].index(el) +2
                    break
            break
    
    wl_l02, wl_l01 = max([med02, m02]), max([med01, m01])

    return wl_l02, wl_l01
```
